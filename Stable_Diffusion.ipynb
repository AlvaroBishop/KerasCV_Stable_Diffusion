```python
import time
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
model = keras_cv.models.StableDiffusion(
    img_width=512, img_height=512, jit_compile=False)
```

```python
prompt = "Rabbit, symmetrical, flat icon design"
```

```python
images = model.text_to_image(prompt, batch_size=3)

def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")


plot_images(images)
```

```python
start = time.time()
images = model.text_to_image(
    "Knolling fruits and vegetables, green background, cartoon style",
    batch_size=3,
)
end = time.time()
plot_images(images)

print(f"Default model: {(end - start):.2f} seconds")
```
